# Employee Attrition Analysis with Sentiment Insights

## Objective
The primary objective of this analysis is to predict employee attrition – identifying whether an employee will leave the company. We aim to explore the potential of combining traditional tabular data with unstructured textual feedback to enhance the predictive power of our models.

## Data Description
The dataset contains conventional, structured employee data alongside a column of unstructured textual feedback. The structured data includes various features like demographics, job characteristics, and performance metrics. The unstructured feedback is in natural language and holds potentially rich insights into employee sentiments.

## Analytical Approach
To achieve our goal, we will train two distinct classification models:
1. **Baseline Model:** Utilizes only the structured information without considering the textual feedback.
2. **Sentiment-Enhanced Model:** Integrates sentiment analysis of the textual feedback into the structured data, hypothesizing that sentiment scores can provide additional context that may correlate with attrition.

## Hypothesis
The sentiment-enhanced model is expected to perform better by leveraging the nuances captured from the employee feedback. The rationale is that employees' sentiments, reflected in their feedback, are indicative of their engagement and satisfaction levels, which in turn could be predictive of their likelihood to stay with or leave the company.
The models' performances will be evaluated and compared using standard classification metrics to understand the impact of incorporating sentiment analysis into attrition prediction.

```python
from sklearn.model_selection import train_test_split

def prepare_data(df, seed):

  # features and target
  X_all = df.drop('Attrition', axis=1)
  y_all = df['Attrition']

  X_no_nlp = df.drop(['Attrition', 'Sentiment_Score', 'Sentiment_Score_Label'], axis=1)
  y_no_nlp = df['Attrition']

  # test set
  X_train_all, X_test_all, y_train_all, y_test_all = train_test_split(X_all, y_all, test_size=0.2, random_state=42)
  X_train_no_nlp, X_test_no_nlp, y_train_no_nlp, y_test_no_nlp = train_test_split(X_no_nlp, y_no_nlp, test_size=0.2, random_state=42)

  # validation set
  X_train_all, X_val_all, y_train_all, y_val_all = train_test_split(X_train_all, y_train_all, test_size=0.25, random_state=42)
  X_train_no_nlp, X_val_no_nlp, y_train_no_nlp, y_val_no_nlp = train_test_split(X_train_no_nlp, y_train_no_nlp, test_size=0.25, random_state=42)

  X_no_nlp = (X_train_no_nlp, X_val_no_nlp, X_test_no_nlp)
  y_no_nlp = (y_train_no_nlp, y_val_no_nlp, y_test_no_nlp)
  X_all = (X_train_all, X_val_all, X_test_all)
  y_all = (y_train_all, y_val_all, y_test_all)

  return X_no_nlp, y_no_nlp, X_all, y_all
```

## Models

```python
X_no_nlp, y_no_nlp, X_all, y_all = prepare_data(df_relevant_all, seed=42)
(X_train_no_nlp, X_val_no_nlp, X_test_no_nlp) = X_no_nlp
(y_train_no_nlp, y_val_no_nlp, y_test_no_nlp) = y_no_nlp
(X_train_all, X_val_all, X_test_all) = X_all
(y_train_all, y_val_all, y_test_all) = y_all
```

```python
class ModelTrainer:
    def __init__(self, input_shape, learning_rate=1e-3, epochs=100, batch_size=32):
        self.input_shape = input_shape
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size
        self.model = self._build_model()

    def _build_model(self):
      model = tf.keras.Sequential([
          tf.keras.layers.Input(shape=self.input_shape),
          tf.keras.layers.Normalization(),
          tf.keras.layers.Dense(units=64, activation="relu"),
          tf.keras.layers.Dropout(rate=0.2),
          tf.keras.layers.Dense(units=128, activation="relu"),
          tf.keras.layers.Dropout(rate=0.2),
          tf.keras.layers.Dense(units=256, activation="relu"),
          tf.keras.layers.Dropout(rate=0.2),
          tf.keras.layers.Dense(units=128, activation="relu"),
          tf.keras.layers.Dropout(rate=0.2),
          tf.keras.layers.Dense(units=64, activation="relu"),
          tf.keras.layers.Dense(units=1, activation="sigmoid")
          ])
      model.compile(
          loss="binary_crossentropy",
          optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
          metrics=[tf.keras.metrics.BinaryAccuracy(), tf.keras.metrics.AUC()]
          )
      return model

    def train(self, X_train, y_train, X_val, y_val, verbose=1):
        history = self.model.fit(
            X_train,
            y_train,
            epochs=self.epochs,
            batch_size=self.batch_size,
            validation_data=(X_val, y_val),
            verbose=verbose
        )
        return history.history

    def evaluate(self, X_test, y_test):
        test_loss, test_accuracy, test_auc = self.model.evaluate(X_test, y_test, verbose=0)
        return test_loss, test_accuracy, test_auc

    def predict(self, X):
        return self.model.predict(X)
```

```python
nlp_model = ModelTrainer(X_train_all.shape[1])
nlp_history = nlp_model.train(X_train_all, y_train_all, X_val_all, y_val_all)
test_loss_nlp, test_accuracy_nlp, test_auc_nlp = nlp_model.evaluate(X_test_all, y_test_all)
print(f"NLP Model Test Accuracy: {100 * test_accuracy_nlp:.4f}%")
print(f"NLP Model Test AUC: {100 * test_auc_nlp:.4f}%")
```

```python
auc_scores_base = []
base_models = []
auc_scores_nlp = []
nlp_models = []

# Number of times to retrain the model
n_runs = 10

for seed in range(n_runs):
    tf.random.set_seed(seed)
    print(f"run #{seed}")

    # Split the data
    X_no_nlp, y_no_nlp, X_all, y_all = prepare_data(df_relevant_all, seed=seed)
    (X_train_no_nlp, X_val_no_nlp, X_test_no_nlp) = X_no_nlp
    (y_train_no_nlp, y_val_no_nlp, y_test_no_nlp) = y_no_nlp
    (X_train_all, X_val_all, X_test_all) = X_all
    (y_train_all, y_val_all, y_test_all) = y_all

    # no nlp model
    base_model = ModelTrainer(X_train_no_nlp.shape[1], epochs=50)
    base_history = base_model.train(X_train_no_nlp, y_train_no_nlp, X_val_no_nlp, y_val_no_nlp, verbose=0)
    _, _, test_auc_base = base_model.evaluate(X_test_no_nlp, y_test_no_nlp)
    auc_scores_base.append(test_auc_base)
    base_models.append(base_model)
    print(f"base auc {100 * test_auc_base:.4f}%")

    # nlp model
    nlp_model = ModelTrainer(X_train_all.shape[1], epochs=50)
    nlp_history = nlp_model.train(X_train_all, y_train_all, X_val_all, y_val_all, verbose=0)
    _, _, test_auc_nlp = nlp_model.evaluate(X_test_all, y_test_all)
    auc_scores_nlp.append(test_auc_nlp)
    nlp_models.append(nlp_model)
    print(f"NLP auc {100 * test_auc_nlp:.4f}%")
    print("-"*50)

# Convert to numpy arrays for statistical testing
auc_scores_base = np.array(auc_scores_base)
auc_scores_nlp = np.array(auc_scores_nlp)
```

```python
from scipy.stats import wilcoxon

# Perform the Wilcoxon signed-rank test
stat, p = wilcoxon(auc_scores_base, auc_scores_nlp) # suggested by chatgpt
print(f'Wilcoxon signed-rank test statistic: {stat}, p-value: {p}')

alpha = 0.05
if p < alpha:
    print('The difference in AUC between the base and NLP models is statistically significant.')
else:
    print('The difference in AUC between the base and NLP models is not statistically significant.')
```

Employee turnover can have a significant effect on the operational efficiency and financial stability of an organization. Thus, the ability to predict and mitigate attrition risks is of relevant importance.

The ML models we built aimed to use the power of sentiment analysis to add a layer of predictive capability to the conventional attrition modeling techniques. However, contrary to our expectations, the inclusion of sentiment analysis-derived features did not lead to a statistically significant improvement in the model's performance.

However, we believe there are several key limitations and areas for further refinement in our approach, especially two:

- Cost Sensitivity: We acknowledged that the costs associated with false positives and false negatives are asymmetric. The financial implications of unexpectedly losing an employee far outweigh the costs of incorrectly predicting an employee's departure. This cost difference necessitates a more nuanced approach to model evaluation, one that goes beyond traditional metrics like accuracy or AUC. In future iterations, we will explore cost-sensitive learning and evaluation techniques that can more accurately reflect the business impact of model predictions.

- Sentiment Analysis Precision: The sentiment analysis utilized in this study was based on a general-purpose model from Hugging Face. While it is a powerful tool, it may not be finely tuned to the nuances of employee feedback in the context of attrition. Tailoring sentiment analysis to capture more specific emotional cues or industry-specific sentiments could yield more predictive features. Further, employing models that can understand context and the subtleties of workplace dynamics could prove more effective.

In conclusion, although the sentiment analysis features did not significantly enhance the model's predictive power in this instance, we remain optimistic about its potential. With targeted improvements and a more refined approach, we believe that sentiment analysis can become a valuable asset in the HR toolkit, aiding in the reduction of attrition rates and leading to considerable cost savings for the company.
